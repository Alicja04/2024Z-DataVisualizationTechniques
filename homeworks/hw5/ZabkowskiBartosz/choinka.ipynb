```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = np.random.uniform(-1, 1, 10000)
y = np.random.uniform(0, 1, 10000)

angles = np.random.uniform(0, 2 * np.pi, 1000)
radii = 0.01

x_circle = 0.3 + radii * np.cos(angles)
y_circle = 0.1 + radii * np.sin(angles)

mask = (np.abs(x) + y) < 1
x_tree = x[mask]
y_tree = y[mask]

x_trunk = np.random.uniform(-0.15, 0.15, 1000)
y_trunk = np.random.uniform(-0.3, 0, 1000)



plt.figure(figsize=(6, 8))
sns.scatterplot(x=x_tree, y=y_tree, color='green', s=10)
sns.scatterplot(x=[0.1], y=[0.1], color='blue', s=70)
sns.scatterplot(x=[-0.1], y=[0.8], color='pink', s=70)
sns.scatterplot(x=[0.6], y=[0.2], color='red', s=70)
sns.scatterplot(x=[0.25], y=[0.6], color='purple', s=70)
sns.scatterplot(x=[-0.25], y=[0.4], color='purple', s=70)
sns.scatterplot(x=[-0.6], y=[0.3], color='white', s=70)
sns.scatterplot(x=[0.25], y=[0.6], color='purple', s=70)
sns.scatterplot(x=[0.05], y=[0.45], color='cyan', s=70)
sns.scatterplot(x=[-0.4], y=[0.1], color='yellow', s=70)
sns.scatterplot(x=[0.3], y=[0.2], color='white', s=70)
sns.scatterplot(x=x_trunk,y=y_trunk, color='brown', s=10)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.xticks([])
plt.yticks([])
plt.savefig('tree.png')
```
